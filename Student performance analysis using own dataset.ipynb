```python
df_cgpa = data['CGPA']
plt.figure(figsize=(9,7))
df_cgpa.plot(style='k.')
plt.title('Scatter plot of cgpa')
plt.show()
```

create new colume of performance score


```python
data['performance_score']=data['CGPA'].shift(-1)
data['performance_score']
```

create new colume for daily return 

```python
data['daily_change']=data['performance_score']-data['CGPA']
data['daily_change']
```

```python
data['ra50']=data['CGPA'].rolling(50).mean()
plt.figure(figsize=(9,7))
data['ra50'].plot()
data['CGPA'].plot()
plt.legend()
plt.show()
```

```python
data['ra50']=data['CGPA'].rolling(50).mean()
data['ra10']=data['CGPA'].rolling(10).mean()
data['ra10'].fillna(1)
```

```python
data=data.dropna()
data
```

```python
# Mark top performers (1 = top performer, 0 = not)
data['Top_Performer'] = (data['CGPA'] > data.groupby('department')['CGPA'].transform('mean')).astype(float)

# Optional: view results
print(data[['student_name', 'department', 'CGPA', 'Top_Performer']])
```

```python
data['CGPA1'] = data['CGPA'].shift(-1)
# Assuming 'Next_CGPA' and 'Top_Performer' already exist
data['CGPA_Gain'] = [data.loc[i, 'CGPA1'] - data.loc[i, 'CGPA']  if data.loc[i, 'Top_Performer']==1 else 1 for i in data.index]

# Plot
data['CGPA_Gain'].plot()

print(data['CGPA1'])
print(data['CGPA_Gain'])
plt.axhline(y=0, color='red')
```

```python
data['log_return']=np.log(data['CGPA'].shift(-1))-np.log(data['CGPA'])
data['log_return']
```

```python
from scipy.stats import norm
mu = data['CGPA'].mean()
sigma = data['CGPA'].std()
prob_return1 = norm.cdf(-0.10, mu, sigma)
print('The probability of CGPA dropping over 10% in one day is', prob_return1)
```

```python
data['log_return'] = pd.to_numeric(data['log_return'], errors='coerce')
data.dropna(subset=['log_return'], inplace=True)

x = np.linspace(data['log_return'].min()-0.01, data['log_return'].max()+0.01, 500)
plt.hist(data['log_return'], bins=50, density=True, alpha=0.6)
plt.plot(x, norm.pdf(x, data['log_return'].mean(), data['log_return'].std()), 'r', lw=2)
plt.show()
```

```python
mu220 = 365*mu
sigma220 = (365**0.5) * sigma
drop20 =  norm.cdf(-0.25, mu220, sigma220)
print('The probability of dropping over 25% over a year: ', drop20)
```
