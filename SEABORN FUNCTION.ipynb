Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x=np.linspace(0,5,11)
y=x**2
print(x)
print(y)
```

Line chart

```python
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title("Line chart")
sns.lineplot(x=x,y=y,color='red')
plt.show()
```

Scatter plot


```python
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title("scatter chart")
sns.scatterplot(x=x,y=y,marker='*',color='green')
plt.show()
```

Multiple line plots

```python

x1 = np.linspace(-np.pi, np.pi, 256)
x2 = np.linspace(-np.pi, np.pi, 256)

y1 = np.cos(x1)
y2 = np.sin(x2)
```

```python
sns.lineplot(x=x1, y=y1,color='green')
sns.lineplot(x=x2,y=y2)
plt.show()
```

Figure Size and dpi

```python
plt.figure(figsize = (10, 6), dpi = 80)
sns.lineplot(x=x1,y= y1)
sns.lineplot(x=x2, y=y2)
plt.show()
```

Styling Figure

```python
plt.figure(figsize = (10, 6), dpi = 80)
sns.lineplot(x=x1, y=y1, color='orange',marker='^')
sns.lineplot(x=x2, y=y2, color='red',marker='*')
plt.show()
```

xticks and yticks

```python
plt.figure(figsize = (10, 6), dpi = 80)
sns.lineplot(x=x1, y=y1, color = 'purple', lw = 2.5, ls = '--')
sns.lineplot(x=x2, y=y2, color = 'green', lw = 2.5, ls = '-.')
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1, 0, 1])
plt.show()
```

```python
plt.figure(figsize = (10, 6), dpi = 80)
sns.lineplot(x=x1, y=y1, color = 'purple', lw = 2.5, ls = '--')
sns.lineplot(x=x2,y= y2, color = 'green', lw = 2.5, ls = '-.')
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], [r'$-\pi$', r'$-\pi/2$', r'$0$',r'$\pi/2$',r'$\pi$'])
plt.yticks([-1, 0, 1])
plt.show()
```

Legend

```python
plt.figure(figsize = (10, 6), dpi = 80)
sns.lineplot(x=x1, y=y1, color = 'purple', lw = 2.5, ls = '--', label = '$y = cosx$')
sns.lineplot(x=x2, y=y2, color = 'green', lw = 2.5, ls = '-.', label = '$y = sinx$')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Graph')
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], [r'$-\pi$', r'$-\pi/2$', r'$0$',r'$\pi/2$',r'$\pi$'])
plt.yticks([-1, 0, 1])
plt.legend(loc = 0)
plt.show()
```

Subplot

```python
plt.figure(figsize = (10, 6), dpi = 80)
plt.subplot(1, 2, 1)
sns.lineplot(x=x1, y=y1, color='r',ls='-.')
plt.subplot(1,2, 2)
sns.lineplot(x=x2,y= y2, color='b',marker='*')
```

```python
fig = plt.figure(figsize = (10, 6))
axes = fig.subplots(2, 3)
```

```python
fig = plt.figure(figsize = (10, 6))
axes = fig.subplots(2, 3)
axes[0][0].plot(x,y,'r--')
axes[0][0].set_title('0,0')
axes[0][1].plot(x,y,'b-.')
axes[0][1].set_title('0,1')
axes[0][2].plot(x,y,'go')
axes[0][2].set_title('0,2')
axes[1,0].plot(y,x,'r--')
axes[1,0].set_title('1,0')
fig.tight_layout()
```

Axes in Figure

```python
fig = plt.figure(figsize = (10,6))
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.10, 0.55, 0.4, 0.3])
axes1.plot(x,y, 'r--')
axes1.set_title('Large Plot')
axes2.plot(y,x, 'g-.')
axes2.set_title('Small Plot')
```

seaborn visualization using dataset

```python
df=sns.load_dataset("fmri")
print(df)
```

line plot

```python
sns.lineplot(x='timepoint',y='signal',data=df)
plt.title("line plot Fmri signal over time")
plt.show()
```

 scatter plot

```python
tips=sns.load_dataset("tips")
print(tips)
```

```python
sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title("scattef plot:Total bill vs Tip")
```

pair plot

```python
iris=sns.load_dataset("iris")
print(iris)
```

```python
sns.pairplot(iris,hue='species')
plt.title("pair plot:Iris dataset")
```

Regression plot

```python
sns.regplot(x='total_bill',y='tip',data=tips)
plt.title("Regression plot:Total boll vs Tip")
```

Heatmap

```python
corr=tips.corr(numeric_only=True)
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.title("Heatmap:correlation matrix of Tips dataset")
```

Box plot

```python
sns.boxplot(data=tips,color='red')
plt.title("Box plot of tips dataset")
```

Histogram

```python
sns.histplot(data=tips,bins=20)
plt.title("Histogram of tip dataset")
```

Bar plot

```python
sns.barplot(x='total_bill',y='tip',data=tips)
plt.title("Bar plot of tip dataset")
```

Count plot

```python
sns.countplot(data=tips)
plt.title("Count plot of tip dataset")
```
