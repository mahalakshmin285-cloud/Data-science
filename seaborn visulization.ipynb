SEABORN THEMES & COLOR PALETTES


```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data=pd.read_csv("c:/Users/User/Downloads/Employee_Salaries.csv") 
print(data)
```

Available Themes

```python
sns.set_theme(style="darkgrid")
sns.set_theme(style="whitegrid")
sns.set_theme(style="dark")
sns.set_theme(style="white")
sns.set_theme(style="ticks")
```


Color Palettes


```python
sns.color_palette("deep")
sns.color_palette("bright")
sns.color_palette("muted")
sns.color_palette("pastel")
sns.color_palette("dark")
sns.color_palette("colorblind")

```

Multi Color Count Plot

```python
sns.set_theme(style="whitegrid")
sns.countplot(data=data,palette="deep")
plt.xticks(rotation=30)
plt.show()
```

Histogram with Theme

```python
plt.figure(figsize = (10, 6), dpi = 80)
sns.set_theme(style="dark")
plt.subplot(1,2, 2)
sns.histplot(x='Overtime_Pay',data=data,palette='bright',bins=10,kde=True)
plt.subplot(1, 2, 1)
sns.histplot(x='Base_Salary',data=data,kde=True,bins=10)
```

Box Plot with themes

```python
sns.set_theme(style="ticks")
sns.boxplot(data=data,color='purple')
```

Heatmap with Themes

```python
corr=data.corr(numeric_only=True)
sns.set_theme(style="darkgrid")
sns.heatmap(corr,annot=True,cmap='Spectral')
```
