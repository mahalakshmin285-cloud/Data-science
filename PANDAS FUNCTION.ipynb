Merging, Joining and Concatenating

```python
df1=pd.DataFrame(np.random.randint(1,20,size=(3,3)),[1,2,3],['maha','sri','bob'])
print(df1)
```

```python
df2 = pd.DataFrame(np.random.randint(20,40, size = (3, 3)), [5,6,7],['maha','sri','bob'])
print(df2)
```

```python
farme=[df1,df2]
pd.concat(farme)
```

```python
pd.concat(farme, axis = 1)
```

merging

```python
left = {
    'key':['k0','k1','k2','k3'],
    'A':[10,20,30,40],
    'B':[1,2,3,4]
}
left = pd.DataFrame(left)
print(left)
```

```python
right = {
    'key':['k0','k1','k2','k3'],
    'C':[100,200,300,400],
    'D':[11,12,13,14]
}
right = pd.DataFrame(right)
print(right)
```

```python
pd.merge(left, right, on='key')
```

joining

```python
left = left.set_index('key')
print(left)
```

```python
right = right.set_index('key')
print(right)
```

```python
result = left.join(right,on='key', how='inner')
print(result)
```
