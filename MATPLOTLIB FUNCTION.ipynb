Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,5,11)
y=x**2
print(x)
print(y)
```

Line chart

```python
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title("Line chart")
plt.plot(x,y,color='red')
plt.show()
```

Scatter plot


```python
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title("scatter chart")
plt.scatter(x,y,marker='*',color='green')
plt.show()
```

Multiple line plots

```python

x1 = np.linspace(-np.pi, np.pi, 256)
x2 = np.linspace(-np.pi, np.pi, 256)

y1 = np.cos(x1)
y2 = np.sin(x2)
```

```python
plt.plot(x1, y1, x2, y2)
plt.show()
```

Figure Size and dpi

```python
plt.figure(figsize = (10, 6), dpi = 80)
plt.plot(x1, y1)
plt.plot(x2, y2)
plt.show()
```

Styling Figure

```python
plt.figure(figsize = (10, 6), dpi = 80)
plt.plot(x1, y1, 'r-.')
plt.plot(x2, y2, 'b--')
plt.show()
```

xticks and yticks

```python
plt.figure(figsize = (10, 6), dpi = 80)
plt.plot(x1, y1, color = 'purple', lw = 2.5, ls = '--')
plt.plot(x2, y2, color = 'green', lw = 2.5, ls = '-.')
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1, 0, 1])
plt.show()
```

```python
plt.figure(figsize = (10, 6), dpi = 80)
plt.plot(x1, y1, color = 'purple', lw = 2.5, ls = '--')
plt.plot(x2, y2, color = 'green', lw = 2.5, ls = '-.')
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], [r'$-\pi$', r'$-\pi/2$', r'$0$',r'$\pi/2$',r'$\pi$'])
plt.yticks([-1, 0, 1])
plt.show()
```

Legend

```python
plt.figure(figsize = (10, 6), dpi = 80)
plt.plot(x1, y1, color = 'purple', lw = 2.5, ls = '--', label = '$y = cosx$')
plt.plot(x2, y2, color = 'green', lw = 2.5, ls = '-.', label = '$y = sinx$')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Graph')
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], [r'$-\pi$', r'$-\pi/2$', r'$0$',r'$\pi/2$',r'$\pi$'])
plt.yticks([-1, 0, 1])
plt.legend(loc = 0)
plt.show()
```

Subplot

```python
plt.figure(figsize = (10, 6), dpi = 80)
plt.subplot(1, 2, 1)
plt.plot(x1, y1, 'r')
plt.subplot(1,2, 2)
plt.plot(x2, y2, 'b-.')
```

```python
fig = plt.figure(figsize = (10, 6))
axes = fig.subplots(2, 3)
```

```python
fig = plt.figure(figsize = (10, 6))
axes = fig.subplots(2, 3)
axes[0][0].plot(x,y,'r--')
axes[0][0].set_title('0,0')
axes[0][1].plot(x,y,'b-.')
axes[0][1].set_title('0,1')
axes[0][2].plot(x,y,'go')
axes[0][2].set_title('0,2')
axes[1,0].plot(y,x,'r--')
axes[1,0].set_title('1,0')
fig.tight_layout()
```

Axes in Figure

```python
fig = plt.figure(figsize = (10,6))
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.10, 0.55, 0.4, 0.3])
axes1.plot(x,y, 'r--')
axes1.set_title('Large Plot')
axes2.plot(y,x, 'g-.')
axes2.set_title('Small Plot')
```

 3D plot

```python
from mpl_toolkits import mplot3d
```

```python
fig = plt.figure(figsize= (10,6))
ax = plt.axes(projection = '3d')
```

3D line plot

```python
fig = plt.figure(figsize= (10,6))
ax = plt.axes(projection = '3d')
zline = np.linspace(0, 15, 1000)
xline = np.sin(zline)
yline = np.cos(zline)
ax.plot3D(xline, yline, zline, 'purple')
ax.set_title("3D line plot")
plt.show()
```

3D scatter plot

```python
fig = plt.figure(figsize= (10,6))
ax = plt.axes(projection = '3d')
zline = np.linspace(0, 15, 100)
xline = np.sin(zline)
yline = np.cos(zline)
ax.scatter(xline, yline, zline, 'yellow')
ax.set_title("3D scatter plot")
plt.show()
```

3D surface plot

```python
x1=np.outer(np.linspace(-2,10),np.ones(10))
y1=x1.copy()
z1=np.cos(x1**2+y1**3)
ax=plt.axes(projection='3d')
ax.plot_surface(x1,y1,z1,cmap='coolwarm',edgecolor='red')
ax.set_title("3d surface plot")
plt.show()
```

3d wireframe plot

```python
def f(x, y):
    return np.sin(np.sqrt(x**2+y**2))
```

```python
x=np.linspace(-1,5,10)
y=x.copy()
X,Y=np.meshgrid(x,y)
z=f(X,Y)
fig=plt.figure()
ax=plt.axes(projection='3d')
ax.plot_wireframe(x,y,z,color='green')
ax.set_title("3D wireframe plot")
plt.show()
```
